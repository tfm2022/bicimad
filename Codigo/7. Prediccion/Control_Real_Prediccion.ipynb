```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# dataComparada = pd.read_csv('../../data/PrediccionComparadaReal.csv')

dataComparada = pd.read_csv('../../data/PrediccionComparadaReal.csv')
```

```python
dataComparada
```

```python
dataEstacion = dataComparada[dataComparada['ESTACION']==49]
```

```python
plt.figure(figsize=(20,8))
plt.plot(dataEstacion['PREDICCION'], label='PredicciÃ³n')
plt.plot(dataEstacion['DEMANDA'], color='red', label='Real')
plt.legend()
```

```python
pd.set_option('display.max_rows', None)
```

```python
cuadro = dataComparada.groupby('ESTACION').agg(REAL=('DEMANDA','sum'), PREDICCION=('PREDICCION','sum'))
cuadro['RELACION'] = np.absolute((cuadro['REAL']-cuadro['PREDICCION'])/cuadro['REAL'])
cuadro.sort_values(by='RELACION')
```
