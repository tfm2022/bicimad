```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd 
import webbrowser
from geopy.distance import geodesic
from geopy.geocoders import Nominatim
app = Nominatim(user_agent="tycgis")
```

```python
madrid_map = folium.Map(location=[40.41831, -3.70275], zoom_start=12)
```

```python
meteo_stations = pd.read_excel("data/stations/Estaciones_control_datos_meteorologicos.xls")
bicimad_stations = pd.read_excel("data/stations/bases_bicimad.xls")
```

```python
latitudes_meteo = meteo_stations['LATITUD'].tolist()
longitudes_meteo = meteo_stations['LONGITUD'].tolist()
etiquetas_meteo = meteo_stations['ESTACIÓN'].tolist()

latitudes_bicimad = bicimad_stations['Latitud'].tolist()
longitudes_bicimad = bicimad_stations['Longitud'].tolist()
etiquetas_bicimad = bicimad_stations['Número'].tolist()
```

```python
locations_meteo = folium.plugins.MarkerCluster().add_to(madrid_map)

# loop through the dataframe and add each data point to the mark cluster
for lat, lon, label in zip(latitudes_meteo,longitudes_meteo,etiquetas_meteo):
        folium.Marker(
        location=[lat, lon],
        icon=folium.Icon(color='red',icon="cloud", prefix='fa'),
        popup=label,
        ).add_to(locations_meteo)

locations_bicimad = folium.plugins.MarkerCluster().add_to(madrid_map)

for lat, lon, label in zip(latitudes_bicimad,longitudes_bicimad,etiquetas_bicimad):
        folium.Marker(
        location=[lat, lon],
        icon=folium.Icon(color='blue',icon="bicycle", prefix='fa'),
        popup=label,
        ).add_to(locations_bicimad)
```
